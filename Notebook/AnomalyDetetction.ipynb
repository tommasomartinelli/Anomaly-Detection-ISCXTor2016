Vedo se ci sono valori NaN nel dataset

```python
nan_values = df.isna().sum()
print(nan_values)
```

Avendo solo due campioni con un campo NaN li considero come valori mancanti, che potrÃ² rimuovere senza perdere informazioni importanti

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
print(df.shape)
```

***Visualizzazione distribuzioni***

```python
num_features = df.select_dtypes(include=['float64', 'int64'])
print(num_features.columns)
cat_features = df.select_dtypes(include=['object'])
print(cat_features.columns)
```

Valutare se rimuovere feature inutili

Visualizzo le distribuzioni

```python
num_features.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
for column in cat_features.columns:
    print(column)
    print(cat_features[column].value_counts())
```
