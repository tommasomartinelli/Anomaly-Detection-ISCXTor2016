Avendo solo due campioni con un campo NaN li considero come valori mancanti, che potrÃ² rimuovere senza perdere informazioni importanti

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
print(df.shape)
print(df.isnull().sum())
```

Ho molti valori possibili per le features categoriche, decidere di utilizzare un one hot encode potrebbe far esplodere la dimensionalitÃ ; opto per la rimozione delle due features in quanto non sono particolarmente importanti ai fini del task, poi rifaccio una valutazione piÃ¹ avanti

```python
df = df.drop(['Source IP', ' Destination IP'], axis=1)
print(df.shape)
```

Boxplot per identificazione degli outlier

```python
plt.figure(figsize=(12, 8))
sns.boxplot(df)
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df.drop(columns=[' Flow Bytes/s']))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Generale con logaritmo
numeric_cols = df.select_dtypes(include=np.number).columns
numeric_cols_noFlow = numeric_cols[numeric_cols != 'Flow Bytes/s']

df_log = df.copy()
df_log[numeric_cols_noFlow] = df_log[numeric_cols_noFlow].apply(lambda x: np.log1p(x) if x.name != 'Flow Bytes/s' else x)

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_log, showfliers=False)
plt.xticks(rotation=45, ha='right')
plt.title('Boxplot delle features su scala logaritmica')
plt.show()
```

```python
tor_data = df[df['label'] == 'TOR']
nontor_data = df[df['label'] == 'nonTOR']

def plot_boxplot(data, label):
    plt.figure(figsize=(16, 6))
    sns.boxplot(data=data.drop(columns=['label']))
    plt.title(f'Boxplot delle features per {label} Traffic')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
plot_boxplot(tor_data, 'Tor')
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=tor_data.drop(columns=[' Flow Bytes/s']))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.title('Boxplot delle features per Traffic Tor')
plt.show()
```

```python
#tor log
tor_data_log = tor_data.copy()
tor_data_log[numeric_cols_noFlow] = tor_data_log[numeric_cols_noFlow].apply(lambda x: np.log1p(x) if x.name != 'Flow Bytes/s' else x)
plt.figure(figsize=(12, 8))
sns.boxplot(data=tor_data_log, showfliers=False)
plt.xticks(rotation=45, ha='right')
plt.title('Boxplot delle features su scala logaritmica per Traffic Tor')
plt.show()
```

```python
plot_boxplot(nontor_data, 'nonTor')
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=nontor_data.drop(columns=[' Flow Bytes/s']))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.title('Boxplot delle features per Traffic NonTor')
plt.show()
```

```python
#nontor log
nontor_data_log = nontor_data.copy()
nontor_data_log[numeric_cols_noFlow] = nontor_data_log[numeric_cols_noFlow].apply(lambda x: np.log1p(x) if x.name != 'Flow Bytes/s' else x)
plt.figure(figsize=(12, 8))
sns.boxplot(data=nontor_data_log, showfliers=False)
plt.xticks(rotation=45, ha='right')
plt.title('Boxplot delle features su scala logaritmica per Traffic NonTor')
plt.show()
```

Gli outliers non sono stati gestiti, verrÃ  valutato in seguito il loro impatto sulle performance dei modelli.

```python
class_distribution = df['label'].value_counts()
print(class_distribution)
```

```python
label_distribution = df['label'].value_counts(normalize=True) * 100
plt.figure(figsize=(8, 6))
sns.countplot(x='label', data=df, hue='label')
plt.title('Distribuzione delle Label')
plt.xlabel('Label')
plt.ylabel('Conteggio')
plt.legend(title='Label', labels=[f'{label} ({percentage:.2f}%)' for label, percentage in zip(label_distribution.index, label_distribution)])
plt.show()
```

Analisi della correlazione

```python
correlation_matrix = df.drop(columns=['label']).corr()
print(correlation_matrix)
```

```python
#Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matrice di Correlazione tra le Features')
plt.show()
```

Ho delle correlazioni NaN, vado ad esplorare le features coinvolte

```python
nan_corr_cols = correlation_matrix.columns[correlation_matrix.isna().all()]
print("Colonne con correlazioni NaN:", nan_corr_cols)
```

```python
unique_values_count = df[nan_corr_cols].nunique()
print("Numero di valori unici per ciascuna colonna coinvolta nelle correlazioni NaN:")
print(unique_values_count.values)
```

Ho due colonne che assumono lo stesso valore per ogni campione del dataset e quindi posso rimuoverle


```python
df = df.drop(columns=nan_corr_cols)
```

In base a quanto visto prima vado a rimuovere features correlate

```python
correlation_matrix = df.drop(columns=['label']).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matrice di Correlazione tra le Features')
plt.show()
```
